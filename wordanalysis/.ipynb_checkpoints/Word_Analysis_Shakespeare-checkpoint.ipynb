### Step 1. Load and process file

```python
import urllib2

url= 'http://www.gutenberg.org/cache/epub/100/pg100.txt'
response = urllib2.urlopen(url)

# print headers from the HTTP server through the info() method
print response.info()
```

```python
# save into file because there is access limit
rawfile = open('shakespeare_raw.txt', 'w')
rawfile.truncate()
for line in response.read():
    rawfile.write(line)
rawfile.close()        
```

```python
# clean up the file, remove the preface information
# select only the first 154 chapters

skip = True
fh = open('shakespeare_raw.txt', 'r')

newfile = open('shakespeare_clean.txt', 'w')
newfile.truncate()

for line in fh.readlines():
        
    if skip:
        if line.startswith('by William Shakespeare'):
            skip = False
        continue
    if line.startswith('THE END'):
        break

    newfile.write(line)
fh.close()            
newfile.close()             
```

### Step 2: Verify the content

```python
# check the content

fh = open('shakespeare_clean.txt')

cnt = 0
for line in fh.readlines():
    line = line.strip()
    cnt += 1
    if cnt < 20:
        # print the first 20 lines
        print line

print "The total lines is {}".format(cnt) 
```

### Step 3. Generate the word cloud image

```python
from wordcloud import WordCloud, STOPWORDS

# read the whole text file.
text = open('shakespeare_clean.txt').read()

# Generate a word cloud image
wordcloud = WordCloud(  max_font_size=60
                      , background_color="white"
                      , stopwords=STOPWORDS).generate(text)

# Display the generated image with matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
