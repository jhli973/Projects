```python
import PyPDF2 as pp
```

```python
test = open('test.pdf', 'rb')
```

```python
pdfReader = pp.PdfFileReader(test)
```

```python
print(pdfReader.numPages)
```

```python
pageObj = pdfReader.getPage(0)
print(pageObj.extractText())
```

```python
test.close()
```

```python
constitution = open('constitution_new.pdf', 'rb')
```

```python
constituReader = pp.PdfFileReader(constitution,strict=False)
```

```python
print(constituReader.numPages)
```

```python
for page in range(constituReader.numPages):
    obj = constituReader.getPage(page)
    print(obj.extractText())
```

```python
constituReader.getPage(0).extractText()
```

```python
constitution.close()
```
