## Parsing ICD codes from web or files

You can look at the file format [Here](http://www.hcup-us.ahrq.gov/toolssoftware/ccs/AppendixASingleDX.txt). 

My goal is to parse all ICD codes for a list of CCS categories and dump them into a table in my SQL server database.

### Define a parsing function

This function takes two arguments: a list of ccs cateegorise and a file

```python
def parsing(categories, fh):
    skip = True
    ccs_category = None
    ccs_description = None
    icd_code = []
    
    for line in fh.readlines():
        line = line.strip()
        line = filter(None, line.split('  '))

        if len(line) > 1 and line[0].isdigit() and not line[1].isdigit() and line[0] in categories:
            ccs_category = line[0]
            ccs_description = line[1]
            skip = False
            continue
        if skip:
            continue
        else:
            if len(line) >0:
                for item in line[0].split(' '):
                    icd_code.append(item)   
        if len(line) < 1:
            print ccs_category, ccs_description
            print icd_code
            print '\n'
            ccs_category = None
            ccs_description = None
            icd_code = []
            skip = True
```

### Write into database

To help achieve this goal, we create a function  'parsingandwrite'  which takes four arguments:
- a categories list
- a file
- codeType: procedure or diagnosis
- planType: planned or potentiallyplanned

```python
def parsingandwrite(categories, fh, codeType, planType):
    skip = True
    ccs_category = None
    ccs_description = None
    icd_code = []
    
    import sys
    import pyodbc
    import ast
    ###create a connection with pyodbc 
    connection_str =    """
    Driver={SQL Server Native Client 11.0};
    Server=localhost;
    Database=test;
    Trusted_Connection=yes;
    """
    db_connection = pyodbc.connect(connection_str)
    db_connection.autocommit = True  #very import, otherwise you will not see the result
    db_cursor = db_connection.cursor()
    
    ### by commentting out the following block, you can create a table here 
    '''
    sql_command =   """
    DROP TABLE CCS_ICDCodes;
    CREATE TABLE CCS_ICDCodes(
    CCS_Category VARCHAR(4),
    CCS_Category_Description VARCHAR(500),
    IcdCode VARCHAR(10),
    Version VARCHAR(10),
    PlanType VARCHAR(50),
    IcdCodeType VARCHAR(20));
    """
    try:
        db_cursor.execute(sql_command)
    except pyodbc.ProgrammingError:
        print("command fail to execute.")
    '''
    
    for line in fh.readlines():
        line = line.strip()
        line = filter(None, line.split('  '))

        if len(line) > 1 and line[0].isdigit() and not line[1].isdigit() and line[0] in categories:
            ccs_category = line[0]
            ccs_description = line[1]
            skip = False
            continue
        if skip:
            continue
        else:
            if len(line) >0:
                for item in line[0].split(' '):
                    sql_command = ("INSERT INTO CCS_ICDCodes (CCS_Category, CCS_Category_Description, IcdCode, Version,PlanType,IcdCodeType) VALUES (?,?,?,?,?,?)")
                    Values = [ccs_category, ccs_description, item, '9', planType, codeType] 
                    db_cursor.execute(sql_command,Values) 
                    
        if len(line) < 1:
            ccs_category = None
            ccs_description = None
            icd_code = []
            skip = True

    db_cursor.close()
    del db_cursor
    db_connection.close()
    print "Done"
```
