## Import Libraries and Modules 

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

%matplotlib inline

from readimagedata import Dataset
```

## Read the Data

```python
file_path = r'C:\Users\dbsnail\ImageFolder\data'
dataset = Dataset()

auth_image_list, unauth_image_list = dataset.traverse_image_dir(file_path)
```

## Check the data 

```python
print("Length of authorized images is {}".format(len(auth_image_list)))
print("Length of unauthorized images is {}".format(len(unauth_image_list)))
```

```python
# we first convert the data into numpy array
authorized_images = np.array(auth_image_list)
unauthorized_images = np.array(unauth_image_list)
```

```python
# show one of the image
img = Image.fromarray(authorized_images[2000,])  
b, g, r = img.split()
img = Image.merge("RGB", (r, g, b))
plt.figure(figsize=(2,3))
plt.imshow(img)
```

```python
img = Image.fromarray(unauthorized_images[2000,])  
b, g, r = img.split()
img = Image.merge("RGB", (r, g, b))
plt.figure(figsize=(2,3))
plt.imshow(img)
```

## Save the Data

```python
from six.moves.urllib.request import urlretrieve
from six.moves import cPickle as pickle


auth_file_name = file_path + '\\' + 'authorized_image_data.pickle'
with open(auth_file_name, 'wb') as af:
    pickle.dump(auth_image_list, af, pickle.HIGHEST_PROTOCOL)
    
unauth_file_name = file_path + '\\' + 'unauthorized_image_data.pickle'
with open(unauth_file_name, 'wb') as uf:
    pickle.dump(unauth_image_list, uf, pickle.HIGHEST_PROTOCOL)
    
af.close()
uf.close()
```
