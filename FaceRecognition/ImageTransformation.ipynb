```python
import cv2
import os
import numpy as np

def tranfom_image(in_path, out_path):
    for file in os.listdir(in_path):
        file_name = os.path.abspath(os.path.join(in_path, file))
        
        image = cv2.imread(file_name)
        
        h, w =image.shape[:2]
        center = (w / 2.0, h / 2.0)
        for angle in range(2, 12, 2):
            M = cv2.getRotationMatrix2D(center, angle, 1.0)
            rotated = cv2.warpAffine(image, M, (w, h))
            new_name = out_path + 'RT'+ str(angle) + file 
            
            cv2.imshow("rotated", rotated)
            cv2.waitKey()
            cv2.imwrite(new_name, rotated)
            
            N = np.float32([[1,0,angle/2], [0,1,angle/2]])
            dst = cv2.warpAffine(image, N, (h, w))
            new_name = out_path + 'ST'+ str(angle) + file 
            cv2.imshow('img', dst)
            cv2.waitKey()
            cv2.imwrite(new_name, rotated)
            k = cv2.waitKey(5) & 0xFF == ord('q')
            if k ==27:
                break
           
                
    cv2.destroyAllWindows()
```

```python
in_path = r'C:\Users\dbsnail\ImageFolder\Data_backup'  
out_path = r'C:\Users\dbsnail\ImageFolder\Data_backup\\'  
tranfom_image(in_path, out_path)
```
